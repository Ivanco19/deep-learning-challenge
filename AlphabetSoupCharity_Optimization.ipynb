# Alphabet Soup Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df_challenge = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df_challenge.head()
```

```python
# Create a copy of application_df_challenge for each model
model1_df = application_df_challenge.copy()
model2_df = application_df_challenge.copy()
model3_df = application_df_challenge.copy()
```

# Model 1:

### Pre-processing

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
model1_df.drop(columns = ['EIN', 'NAME'], axis=1, inplace=True)

# APPLICATION TYPES
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T9', 'T13', 'T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']

# Replace in dataframe
model1_df['APPLICATION_TYPE'] = model1_df['APPLICATION_TYPE'].apply(lambda x: 'Other' if x in application_types_to_replace else x)

# CLASSIFICATIONS
# use the variable name `classifications_to_keep`
classifications_to_keep = ['C1000', 'C2000', 'C1200', 'C3000', 'C2100']

# Replace in the dataframe
model1_df['CLASSIFICATION'] = model1_df['CLASSIFICATION'].apply(lambda x: 'Other' if x not in classifications_to_keep else x)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies1 = ['AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'APPLICATION_TYPE', 'SPECIAL_CONSIDERATIONS']
model1_df = pd.get_dummies(model1_df, columns = dummies1, dtype=int)
```

### Compile, Train and Evaluate the Model

```python
# Split our preprocessed data into our features and target arrays
X1 = model1_df.drop(columns= ['IS_SUCCESSFUL'])
y1 = model1_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2)

# Create a StandardScaler instances
scaler1 = StandardScaler()

# Scale the data
X_train_scaled1 = scaler1.fit_transform(X_train1)
X_test_scaled1 = scaler1.transform(X_test1)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn1 = tf.keras.models.Sequential()

# First hidden layer
nn1.add(tf.keras.layers.Dense(units = 40, activation = "tanh", input_dim = X1.shape[1]))

# Second hidden layer
nn1.add(tf.keras.layers.Dense(units = 20, activation = 'tanh'))

# Output layer
nn1.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Compile the model
nn1.compile(loss="binary_crossentropy", optimizer='Adam', metrics=["accuracy"])

# Train the model
model1 = nn1.fit(X_train_scaled1, y_train1, epochs=150)
```

```python
# Evaluate the model using the test data
model1_loss, model1_accuracy = nn1.evaluate(X_test_scaled1, y_test1, verbose=2)
print(f"Loss: {model1_loss}, Accuracy: {model1_accuracy}")
```

# Model 2:

### Pre-processing

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
model2_df.drop(columns = ['EIN', 'NAME', 'SPECIAL_CONSIDERATIONS', 'STATUS'], axis=1, inplace=True)
```

```python
# use the variable name `application_types_to_replace`
application_to_keep = ['T9', 'T13', 'T12', 'T14', 'T29', 'T15', 'T17', 'T5', 'T6', 'T10', 'T2', 'T25']

# Replace in the dataframe
model2_df['APPLICATION_TYPE'] = model2_df['APPLICATION_TYPE'].apply(lambda x: 'Other' if x in application_to_keep else x)
model2_df['APPLICATION_TYPE'].value_counts()
```

```python
# use the variable name `classifications_to_keep`
classifications_to_keep = ['C1000', 'C2000', 'C1200', 'C3000', 'C2100']

# Replace in the dataframe
model2_df['CLASSIFICATION'] = model2_df['CLASSIFICATION'].apply(lambda x: 'Other' if x not in classifications_to_keep else x)
model2_df['CLASSIFICATION'].value_counts()
```

```python
# Replace in the dataframe values greater than 100K USD for ASK_AMT
model2_df['ASK_AMT'] = model2_df['ASK_AMT'].apply(lambda x: 'Less than 100K' if x <= 100000 else 'More than 100K')
model2_df['ASK_AMT'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies2 = ['AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'APPLICATION_TYPE', 'ASK_AMT']
model2_df = pd.get_dummies(model2_df, columns = dummies2, dtype=int)
```

```python
# Split our preprocessed data into our features and target arrays
X2 = model2_df.drop(columns= ['IS_SUCCESSFUL'])
y2 = model2_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2)
```

### Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(tf.keras.layers.Dense(units = 128, activation = "relu", input_dim = X2.shape[1]))

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units = 64, activation = 'relu'))

# Third hidden layer
nn2.add(tf.keras.layers.Dense(units = 32, activation = 'relu'))

# Output layer
nn2.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Compile the model
nn2.compile(loss="binary_crossentropy", optimizer="Adam", metrics=["accuracy"])

# Train the model
model2 = nn2.fit(X_train2, y_train2, epochs=200)
```

```python
# Evaluate the model using the test data
model2_loss, model2_accuracy = nn2.evaluate(X_test2, y_test2, verbose=2)
print(f"Loss: {model2_loss}, Accuracy: {model2_accuracy}")
```

# Model 3

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
model3_df.drop(columns = ['EIN', 'NAME'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder instance
onehot_encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the OneHotEncoder using the categorical variable list
df_encoded = pd.DataFrame(onehot_encoder.fit_transform(model3_df[['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS', 'STATUS']]))
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
numeric_column_scaled = scaler.fit_transform(model3_df[['ASK_AMT']])

# Add the numeric_column_scaled to df_encoded dataframe
df_encoded['numeric_column'] = numeric_column_scaled

# Add the IS_SUCCESSFUL column to df_encoded dataframe
df_encoded['IS_SUCCESSFUL'] = model3_df["IS_SUCCESSFUL"]
df_encoded
```

```python
# Split our preprocessed data into our features and target arrays
X3 = df_encoded.drop(columns= ['IS_SUCCESSFUL'])
y3 = df_encoded['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2)
```

### Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn3 = tf.keras.models.Sequential()

# First hidden layer
nn3.add(tf.keras.layers.Dense(units = 128, activation = "relu", input_dim = X3.shape[1]))

# Second hidden layer
nn3.add(tf.keras.layers.Dense(units = 64, activation = 'relu'))

# Third hidden layer
nn3.add(tf.keras.layers.Dense(units = 32, activation = 'relu'))

# Output layer
nn3.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Compile the model
nn3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model3 = nn3.fit(X_train3, y_train3, epochs = 50)
```

```python
# Evaluate the model using the test data
model3_loss3, model3_accuracy3 = nn3.evaluate(X_test3, y_test3, verbose=2)
print(f"Loss: {model3_loss3}, Accuracy: {model3_accuracy3}")
```
